## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the anonymized starter dataset
df = pd.read_csv('content_refresh_anonymized.csv')

# 1. Total available content candidates in the starter slice
total_pages = len(df)

# 2. How many of these are actively showing downward trends?
declining_pages = len(df[df['trend_direction'] == 'down'])

# 3. Filtering for minimum volume (e.g., at least 500 impressions) to strip out noise
high_volume_declining = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] >= 500)])

print(f"Total pages in starter slice: {total_pages:,}")
print(f"Pages with an observed decline: {declining_pages:,}")
print(f"High-volume declining pages (actionable targets): {high_volume_declining:,}")
```
