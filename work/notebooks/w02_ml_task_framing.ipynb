## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule (like `IF ctr < 1.5 AND impressions > 500 THEN decline`) is too rigid and brittle for this dataset. 

1. **Varying Missingness:** Missing data follows content types (e.g., keyword data might be 100% missing for some formats). A fixed rule breaks or requires endless nested logic to handle this, whereas a tree-based model (or even logistic regression, with engineered missingness flags like `has_keyword_data`) adapts naturally.
2. **Client Variability:** We are scoring across 32 different pseudonymized clients. What constitutes "good" engagement or CTR for one client's niche might be terrible for another. 
3. **Tangled Signals:** Metrics like `scroll_rate`, `ai_traffic_pct`, and `avg_position` interact in non-linear ways. ML can weigh these complex interactions simultaneously to find genuine decline signals that simple if-statements would miss.
